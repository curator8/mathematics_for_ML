## dot product  

```python
v1 = np.array([1,2,3,4,5])
v2 = np.array([4,5,6,7,8])


# method 1
dp1 = sum(np.multiply(v1,v2))

# method 2 dot product 
dp2 = np.dot(v1, v2)

# method 3 using matrix multiplication 
dp3 = np.matmul(v1,v2)

#loop version 
dp4 = 0

for i in range(0,len(v1)):
    
    dp4 = dp4 + v1[i] * v2[i]
    

print(dp1,dp2,dp3,dp4)


```
